```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
weather_data = pd.read_csv("data/external_data.csv")
```

```python
threshold = 0.7 * len(weather_data)
weather_data = weather_data.loc[:, weather_data.isnull().sum() <= threshold]

# Fill missing values with median for numeric columns
weather_data = weather_data.fillna(weather_data.median(numeric_only=True))

# Handle categorical columns by replacing missing values with mode
categorical_columns = weather_data.select_dtypes(include=['object', 'category']).columns
for col in categorical_columns:
    weather_data[col] = weather_data[col].fillna(weather_data[col].mode()[0])

columns_to_keep = [
    "ff",
    "t",
    "u",
    "ww",
    "n",
    "etat_sol",
    "ht_neige",
    "rr1",
    "rr3",
    "rr6",
    "rr12",
    "rr24",
]

weather_data = weather_data[columns_to_keep]
```

```python
if 'date_column' in weather_data.columns:
    weather_data['date_column'] = pd.to_datetime(weather_data['date_column'])
    weather_data['date_numeric'] = weather_data['date_column'].map(lambda x: x.timestamp())

heatmap_data = weather_data[[col for col in columns_to_keep if col in weather_data.select_dtypes(include=[np.number]).columns]]
correlation_matrix = heatmap_data.corr()

plt.figure(figsize=(15, 12))
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap="coolwarm", 
    fmt=".2f", 
    annot_kws={"size": 8},
    cbar_kws={"shrink": 0.8}
)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
